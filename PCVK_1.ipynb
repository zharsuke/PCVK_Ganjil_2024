```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io, transform
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# create url lists
urls = ["https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSfPadMIsHgHdhz8hO_WCoLj-aedbKy4UJRlg&s",
        "https://media.zenfs.com/en/blavity_articles_483/d4f5f69205730c9542c8c3e42f4e3f65",
        "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTS7wEd5eOpm0vJMOdvsW1qtrj16PBZ2Jfusg&s"]

# loop through the urls and display the images
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print("\n")
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Tinggi: ", tinggi)
print("Lebar: ", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#horizontal
for y in range(lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

cv2_imshow(image_3)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#vertical
for x in range(tinggi):
  image_3[x,int((lebar)/2)] = [255,255,255]

cv2_imshow(image_3)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Draw diagonal line from top-left to bottom-right
for i in range(min(tinggi, lebar)):
  image_3[i, i] = [255, 255, 255]

# Draw diagonal line from top-right to bottom-left
for i in range(min(tinggi, lebar)):
  image_3[i, lebar - i - 1] = [255, 255, 255]

cv2_imshow(image_3)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#horizontal
for y in range(lebar):
  if (y > lebar / 4) and (y < 3 * lebar / 4):
    image_3[int((tinggi)/2),y] = [255,255,255]

cv2_imshow(image_3)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menentukan area kotak
kotak_x_start = int(lebar / 4)
kotak_x_end = int(3 * lebar / 4)
kotak_y_start = int(tinggi / 4)
kotak_y_end = int(3 * tinggi / 4)

# Menggambar kotak penuh dengan for loop
for x in range(kotak_y_start, kotak_y_end):
    for y in range(kotak_x_start, kotak_x_end):
        image_3[x, y] = [255, 255, 255]  # Warna kotak (putih)

# Menampilkan hasil
cv2_imshow(image_3)
```

```python
# Link gambar KTP
link = "https://fahum.umsu.ac.id/blog/wp-content/uploads/2024/06/cara-dan-syarat-pembuatan-ktp.webp"
ktp = io.imread(link)
ktp = cv.resize(ktp, (0,0), fx=0.5, fy=0.5)
ktp_2 = cv.cvtColor(ktp, cv.COLOR_BGR2RGB)

# Area sensor
sensor_width = 100
sensor_height = 10
sensor_color = [255, 255, 255]  # Warna sensor (putih)

# Menentukan posisi sensor
sensor_x_start = 100
sensor_x_end = sensor_x_start + sensor_width
sensor_y_start = 50
sensor_y_end = sensor_y_start + sensor_height

# Menggambar sensor
for x in range(sensor_x_start, sensor_x_end):
    for y in range(sensor_y_start, sensor_y_end):
        ktp_2[y, x] = sensor_color

# Area sensor
sensor_width2 = 75
sensor_height2 = 10
sensor_color2 = [255, 0, 255]  # Warna sensor (putih)

# Menentukan posisi sensor
sensor_x_start2 = 100
sensor_x_end2 = sensor_x_start2 + sensor_width2
sensor_y_start2 = 60
sensor_y_end2 = sensor_y_start2 + sensor_height2

# Menggambar sensor
for x in range(sensor_x_start2, sensor_x_end2):
    for y in range(sensor_y_start2, sensor_y_end2):
        ktp_2[y, x] = sensor_color2

# Menampilkan hasil
cv2_imshow(ktp_2)
```
